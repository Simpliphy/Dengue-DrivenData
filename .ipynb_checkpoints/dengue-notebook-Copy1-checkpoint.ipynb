# Apprentissage #
----------------------------------------

 ## Ville SJ ##

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn import metrics

```

```python
X_sj = (features_dengue.drop(["week_start_date"],axis=1))
X_sj = X_sj.loc[X_sj['city'] =="sj"].drop(['city'], axis=1)
X_sj = X_sj.fillna(X_sj.mean())

y_sj = labels_dengue.loc[labels_dengue['city'] =="sj"].drop(['city'], axis=1)
y_sj = y_sj["total_cases"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(preprocessing.scale(X_sj), y_sj, test_size=0.3, random_state=101)
reg = linear_model.SGDRegressor()
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
param_grid = {
  'alpha': [0.1, 1, 10], 'epsilon' : [0.01, 0.1, 1.0] ,'n_iter':[5,10,20,50]}
```

```python
grid_sj = GridSearchCV(estimator=linear_model.SGDRegressor(), param_grid=param_grid)
grid_sj.fit(X_train, y_train)
```

```python
y_pred = grid_sj.predict(X_test)
```

```python
sns.distplot((y_pred),bins=50);
```

## score de performance ##

```python
mean_squared_error(y_test, y_pred)  
```

```python
r2_score(y_test, y_pred) 
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
sns.distplot((y_test-y_pred),bins=50);
```

# Ville iq #

```python
X_iq = (features_dengue.drop(["week_start_date"],axis=1))
X_iq = X_iq.loc[X_iq['city'] =="iq"].drop(['city'], axis=1)
X_iq = X_iq.fillna(X_sj.mean())

y_iq = labels_dengue.loc[labels_dengue['city'] =="iq"].drop(['city'], axis=1)
y_iq = y_iq["total_cases"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(preprocessing.scale(X_iq), y_iq, test_size=0.3, random_state=101)

param_grid = {
  'alpha': [0.1, 1, 10], 'epsilon' : [0.01, 0.1, 1.0] ,'n_iter':[5,10,20,50]}

grid_iq = GridSearchCV(estimator=linear_model.SGDRegressor(), param_grid=param_grid)
grid_iq.fit(X_train, y_train)
```

```python
y_pred = grid_iq.predict(X_test)
```

## Score de performance ## 

```python
sns.distplot((y_pred),bins=50);
```

```python
mean_squared_error(y_test, y_pred)  
r2_score(y_test, y_pred) 
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
sns.distplot((y_test-y_pred),bins=50);
```

```python
features_submission.info()
```

## prediction pour ville San Jose ##

```python
features_submission_sj = (features_submission.drop(["week_start_date"],axis=1))
features_submission_sj = features_submission_sj.loc[features_submission_sj['city'] =="sj"].drop(['city'], axis=1)
features_submission_sj = features_submission_sj.fillna(features_submission_sj.mean())
```

```python
predictions_submission_sj = grid_sj.predict(preprocessing.scale(features_submission_sj))
```

## prediction pour ville IQ ##

```python
features_submission_iq = (features_submission.drop(["week_start_date"],axis=1))
features_submission_iq = features_submission_iq.loc[features_submission_iq['city'] =="iq"].drop(['city'], axis=1)
features_submission_iq = features_submission_iq.fillna(features_submission_iq.mean())
```

```python
predictions_submission_iq = grid_iq.predict(preprocessing.scale(features_submission_iq))
```

## Merge des resultats pour soumission et enregistrement ##

```python
from math import floor
```

```python
df2 = pd.concat([pd.DataFrame(predictions_submission_sj), pd.DataFrame(predictions_submission_iq)], axis=0)
df2[0] = df2[0].apply(floor)
df2[0] = df2[0].apply(floor)
print(df2.shape)
print(features_submission.shape)
```

```python

df1 = features_submission[["city", "year", "weekofyear"]]
df1.reset_index(drop=True, inplace=True)
df2.reset_index(drop=True, inplace=True)

results = pd.concat( [df1, df2], axis=1) 
```

```python
results.reset_index(drop=True, inplace=True)
results.columns
results.rename(columns={0: 'total_cases'}, inplace=True)
```

```python
results.to_csv("results.csv",index=False,index_label=False)
```
